**Ejercicio 1: Exploración de Datos Financieros**

Supongamos que tenemos un conjunto de datos de empleados con la siguiente información: cuadro.

**Instrucciones:**
1. Crea un DataFrame en Python con estos datos.
2. Realiza un análisis univariado sobre la variable "Salario" (media, mediana, histograma y boxplot).
3. Calcula la correlación entre Edad y Salario.
4. Visualiza la relación con un diagrama de dispersión.

```python
#1. Crea un DataFrame en Python con estos datos.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creación del  DataFrame
df = pd.DataFrame({
 "Empleado": [1, 2, 3, 4, 5],
 "Edad": [25, 40, 35, 50, 30],
 "Salario": [50000, 60000, 70000, 80000, 45000],
 "Experiencia": [2, 10, 7, 15, 3]
})
```

```python
# Análisis Univariado: Estadísticas básicas
# 2. Realiza un análisis univariado sobre la variable "Salario" (media, mediana, histograma y boxplot).
print(df["Salario"].describe())

# Histograma de salarios
sns.histplot(df["Salario"], bins=5, kde=True)
plt.title("Distribución de Salarios")
plt.show()

# Boxplot para detectar outliers en Salario
sns.boxplot(x=df["Salario"])
plt.title("Boxplot de Salarios")
plt.show()
```

```python
#3. Calcula la correlación entre Edad y Salario.
# Análisis Multivariado: Correlación entre Edad y Salario
print(df[["Edad", "Salario"]].corr())
```

```python
#4. Visualiza la relación con un diagrama de dispersión.
# Gráfico de dispersión
sns.scatterplot(x=df["Edad"], y=df["Salario"])
plt.title("Relación entre Edad y Salario")
plt.show()
```

#**Ejercicio para la Sección 3**

**Objetivo:** Aplicar PCA y Clustering en un conjunto de datos real.

**Ejercicio 1: Reducción de Dimensionalidad y Agrupamiento**

Dado el siguiente conjunto de datos de clientes: cuadro.

**Instrucciones:**
1. Normaliza los datos antes de aplicar PCA.
2. Aplica PCA para reducir a 2 componentes principales y visualiza los resultados.
3. Aplica K-Means Clustering con 2 clusters y grafica los resultados.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Crear DataFrame
df = pd.DataFrame({
 "Cliente": [1, 2, 3, 4, 5],
 "Edad": [25, 40, 35, 50, 30],
 "Ingresos": [50000, 60000, 70000, 80000, 45000],
 "Experiencia": [2, 10, 7, 15, 3]
})

# Normalización
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[["Edad", "Ingresos", "Experiencia"]])

# Aplicar PCA
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)
df_pca = pd.DataFrame(df_pca, columns=["PC1", "PC2"])

# Aplicar K-Means Clustering
kmeans = KMeans(n_clusters=2, random_state=42)
df_pca["Cluster"] = kmeans.fit_predict(df_pca)

# Visualizar clusters
sns.scatterplot(x=df_pca["PC1"], y=df_pca["PC2"], hue=df_pca["Cluster"], palette="viridis")
plt.title("Clusters después de PCA")
plt.show()
```
