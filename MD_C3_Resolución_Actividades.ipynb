**Ejercicio 1: Exploración de Datos Financieros**

Supongamos que tenemos un conjunto de datos de empleados con la siguiente información: cuadro.

**Instrucciones:**
1. Crea un DataFrame en Python con estos datos.
2. Realiza un análisis univariado sobre la variable "Salario" (media, mediana, histograma y boxplot).
3. Calcula la correlación entre Edad y Salario.
4. Visualiza la relación con un diagrama de dispersión.

```python
#1. Crea un DataFrame en Python con estos datos.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creación del  DataFrame
df = pd.DataFrame({
 "Empleado": [1, 2, 3, 4, 5],
 "Edad": [25, 40, 35, 50, 30],
 "Salario": [50000, 60000, 70000, 80000, 45000],
 "Experiencia": [2, 10, 7, 15, 3]
})
```

```python
# Análisis Univariado: Estadísticas básicas
# 2. Realiza un análisis univariado sobre la variable "Salario" (media, mediana, histograma y boxplot).
print(df["Salario"].describe())

# Histograma de salarios
sns.histplot(df["Salario"], bins=5, kde=True)
plt.title("Distribución de Salarios")
plt.show()

# Boxplot para detectar outliers en Salario
sns.boxplot(x=df["Salario"])
plt.title("Boxplot de Salarios")
plt.show()
```

```python
#3. Calcula la correlación entre Edad y Salario.
# Análisis Multivariado: Correlación entre Edad y Salario
print(df[["Edad", "Salario"]].corr())
```

```python
#4. Visualiza la relación con un diagrama de dispersión.
# Gráfico de dispersión
sns.scatterplot(x=df["Edad"], y=df["Salario"])
plt.title("Relación entre Edad y Salario")
plt.show()
```

#**Ejercicio para la Sección 3**

**Objetivo:** Aplicar PCA y Clustering en un conjunto de datos real.

**Ejercicio 1: Reducción de Dimensionalidad y Agrupamiento**

Dado el siguiente conjunto de datos de clientes: cuadro.

**Instrucciones:**
1. Normaliza los datos antes de aplicar PCA.
2. Aplica PCA para reducir a 2 componentes principales y visualiza los resultados.
3. Aplica K-Means Clustering con 2 clusters y grafica los resultados.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Crear DataFrame
df = pd.DataFrame({
 "Cliente": [1, 2, 3, 4, 5],
 "Edad": [25, 40, 35, 50, 30],
 "Ingresos": [50000, 60000, 70000, 80000, 45000],
 "Experiencia": [2, 10, 7, 15, 3]
})

# Cluster de datos originales
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10) # Added n_init to avoid warning
df["Cluster"] = kmeans.fit_predict(df[["Edad", "Ingresos", "Experiencia"]])

# Visualizar clusters con datos originales
sns.scatterplot(x=df["Edad"], y=df["Ingresos"], hue=df["Cluster"], palette="Set1")
plt.title("Clusters Datos Original")
plt.show()

# Normalización
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[["Edad", "Ingresos", "Experiencia"]])

# Aplicar PCA
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)
df_pca = pd.DataFrame(df_pca, columns=["PC1", "PC2"])

print("\n=== VARIANZA EXPLICADA ===")
print(f"PC1: {pca.explained_variance_ratio_[0]*100:.2f}%")
print(f"PC2: {pca.explained_variance_ratio_[1]*100:.2f}%")
print(f"Total: {sum(pca.explained_variance_ratio_)*100:.2f}%")

# Aplicar K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10) # Added n_init to avoid warning
df_pca["Cluster"] = kmeans.fit_predict(df_pca)

# Visualizar clusters
sns.scatterplot(x=df_pca["PC1"], y=df_pca["PC2"], hue=df_pca["Cluster"], palette="Set1")
plt.title("Clusters después de PCA")

# Etiquetas de cliente
for i in range(len(df_pca)):
    plt.text(df_pca["PC1"][i]+0.03, df_pca["PC2"][i], f'C{df["Cliente"][i]}') # Corrected: Access 'Cliente' from original df
plt.show() # Added plt.show() to display the plot with labels

for i in range(3):
    print(f"\nCluster {i}")
    print(df[df['Cluster'] == i])
```

**Preguntas de Reflexión**

1. ¿Cuánta variabilidad explican las dos primeras componentes principales?

**Las dos primeras componentes principales explican el 99,78% de la variabilidad total de los datos. Esto indica que gran parte de la información original puede representarse en solo dos dimensiones.**

2. ¿Cómo cambian los resultados de clustering después de aplicar PCA?

**Después de aplicar PCA, se obtuvieron dos clusteres bien diferenciados y dentro de cada uno se observa una gran distancia, que deberá analizarse.**

3. ¿Qué conclusiones puedes extraer de los clusters obtenidos?

**Se obtuvo una segmentación de los clientes en tres grupos distintos, segúnsu edad, ingresos y experiencia.**

**Cluster 0 (Clientes 2 y 3): son clientes de edad y experiencia intermedia (entre 35 y 40 años, con 7 a 10 años de experiencia) y salarios en un rango medio-alto ($60.000 - $70.000).**

**Cluster 1 (Cliente 4): Este cluster es un segmento de clientes de alto valor, caracterizado por tener la mayor edad (50 años), los ingresos más altos ($80.000) y la mayor experiencia (15 años). Representa a clientes con una trayectoria consolidada.**

**Cluster 2 (Clientes 1 y 5): representa a los clientes más jóvenes (25 y 30 años) con menos experiencia (2 y 3 años) y los salarios más bajos ($45.000 - $50.000). Podrían ser clientes en las primeras etapas de su carrera.**
