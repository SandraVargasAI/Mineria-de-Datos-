# Se importan las librerías necesarias para la manipulación de datos y para generar los gráficos.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configuración global de estilo
sns.set_theme(style="whitegrid", palette="Set2")
```

# Crear DataFrame con losdatos que vamos  utilizar (cliente, edad, ingresos, compras y categoría).


```python
df = pd.DataFrame({
    "Cliente":   [1, 2, 3, 4, 5, 6, 7, 8],
    "Edad":      [25, 40, 35, 50, 30, 45, 38, 29],
    "Ingresos":  [50000, 60000, 70000, 80000, 45000, 85000, 72000, 48000],
    "Compras":   [5, 15, 10, 20, 7, 18, 12, 6],
    "Categoría": ["A", "B", "C", "A", "B", "C", "A", "B"]
})
print(df)
```

# **1) Análisis Exploratorio**
Mediante los gráficos vamos a analizar como se distribuyen las variables y veremos si hay alguna relación entre ellas.

```python
# 1) Análisis Exploratorio
# Múltiples sucplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Subplot1: Histograma de Edad
sns.histplot(df["Edad"], bins=5, kde=True, ax=axes[0])
axes[0].set_title("Distribución de Edades")
axes[0].set_xlabel("Edad")
axes[0].set_ylabel("Frecuencia")

# Subplot2
sns.histplot(df["Ingresos"], bins=5, kde=True, ax=axes[1])
axes[1].set_title("Distribución de Ingresos")
axes[1].set_xlabel("Ingresos")
axes[1].set_ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

```python
# Boxplot de Compras
sns.boxplot(x=df["Compras"])
plt.title("Boxplot de Compras")
plt.xlabel("Cantidad de Compras")
plt.show()
```

```python
#Scatterplot: Ingresos y Compras
sns.scatterplot(x="Ingresos", y="Compras", hue="Categoría", data=df, s=100)
plt.title("Relación entre Ingresos y Compras")
plt.xlabel("Ingresos")
plt.ylabel("Compras")
plt.show()
```

# **2) Visualización Avanzada**

```python
# Vemos la relación entre las distintas  variable

sns.pairplot(df[["Edad", "Ingresos", "Compras"]])
plt.show()
```

```python
# Mapa de calor de correlaciones, entre variables numéricas

correlaciones = df.corr(numeric_only=True)
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación de Variables")
plt.show()
```

```python
# Violin Plot: distribución de Compras por Categoría

sns.violinplot(x="Categoría", y="Compras", data=df, hue="Categoría", palette="muted", legend=False)
plt.title("Distribución de Compras por Categoría")
plt.xlabel("Categoría")
plt.ylabel("Compras")
plt.show()
```

# Conclusión

En base a los gráficos, podemos ver que hay una relación positiva entre los ingresos y compras, personas con mayores ingresos suelen comprar más. Contrario a las personas con menor ingreso que compran menos. También se observa que la edad, se asocia a mayores ingresos, por lo tanto realizan mayores compras. En fin, las variables ingresos y edad, determinan el comportamiento de lascompras.

### Gráfico de Burbujas: Ingresos, Compras, Edad y Categoría

Vamos a crear un gráfico de burbujas para visualizar la relación entre los ingresos, la cantidad de compras y la edad de los clientes, diferenciando por categoría. Este gráfico nos permitirá identificar visualmente patrones en la rentabilidad de los clientes.

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x="Ingresos",
    y="Compras",
    size="Edad",
    hue="Categoría",
    data=df,
    sizes=(50, 1000),
    alpha=0.7
)
plt.title("Gráfico de Burbujas: Ingresos, Compras y Edad por Categoría")
plt.xlabel("Ingresos")
plt.ylabel("Compras")
plt.legend(title="Categoría", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
