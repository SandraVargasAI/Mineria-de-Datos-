```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Comparara el salario promedio en diferentes industrias.

```python
df = pd.DataFrame({
    "Industria": [
        "Tecnología", "Finanzas", "Salud", "Educación", "Manufactura",
        "Pyme", "Construcción", "Marketing", "Hotelería", "Transporte"
    ],
    "Salario_Promedio": [
        85000, 92000, 78000, 60000, 68000,
        55000, 62000, 70000, 48000, 65000
    ]
})
print(df)
```

Utilizo un Gráfico de Barras, ya que me permite representar dos variables según su categoría (Industria) y valor (Salario).

```python
sns.barplot(x="Industria", y="Salario_Promedio", data=df)
plt.title("Salario Promedio por Industria")
plt.xlabel("Industria")
plt.ylabel("Salario Promedio")
plt.xticks(rotation=45, ha='right') # Rota las etiquetas del eje X
plt.tight_layout() # Ajusta automáticamente el diseño para evitar superposiciones
plt.show()
```

# Analizar si existe una relación entre la cantidad de horas trabajadas y el nivel de productividad.

```python
df1 = pd.DataFrame({
    "Horas_Trabajadas": [1, 3, 5, 8, 9, 10],
    "Nivel_Productividad": [2, 4, 6, 8, 10, 12]
})
```

Utilizo un gráfico de dispersión (Scatter Plot). Podemos ver si hay una relación entre las dos varibles numéricas de estudio. Para determinar si en el caso de aumentar la cantidad de horas trabajadas, se observa un aumento o dismunuye o se mantiene el nivel de productividad.

```python
sns.scatterplot(x="Horas_Trabajadas", y="Nivel_Productividad", data=df1)
plt.title("Relación entre Horas Trabajadas y Nivel de Productividad")
plt.xlabel("Horas Trabajadas")
plt.ylabel("Nivel de Productividad")
plt.show()
```

# Detectar valores atípicos en los montos de transacciones bancarias.

```python
df2 = pd.DataFrame({
    "Monto": [100, 80, 150, 1000, 35, 60, 400, 350]
})
```

Para detectar valores átipicos, utilizamos el gráfico Boxplot. Además nos brinda información respecto a la mediana y la dispersión de los datos.

```python
sns.boxplot(x=df2["Monto"])
plt.title("Boxplot de Montos de Transacciones Bancarias")
plt.xlabel("Monto")
plt.show()
```

# Observar la distribución de notas de un curso.

```python
df3 = pd.DataFrame({
    "Nota": [10, 5, 6, 4, 2, 8, 10, 7, 6, 8, 10, 8, 7]
})
```

En estecaso vamos a utilizar un Histograma, para observar ladistribución de la variable numérica (Nota).

```python
sns.histplot(df3["Nota"], bins=5, kde=True)
plt.title("Distribución de Notas de un Curso")
plt.xlabel("Nota")
plt.ylabel("Frecuencia")
plt.show()
```
