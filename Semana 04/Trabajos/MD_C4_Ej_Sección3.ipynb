```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Carga el dataset de diamantes de Seaborn (diamonds).

```python
# Cargar dataset
df = sns.load_dataset("diamonds")
df.head()
```

# Crea un Pairplot para analizar la relación entre precio, quilates y profundidad.

```python
# 1. Pairplot entre precio, quilates y profundidad
sns.pairplot(df[["price", "carat", "depth"]])
plt.show()
```

# Genera un Heatmap para visualizar la correlación entre las variables numéricas.

```python
# 2. Heatmap de correlaciones
correlaciones = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación de Diamantes")
plt.show()
```

# Crea un Violin Plot para comparar la distribución del precio en cada categoría de corte (cut).

```python
# 3. Violin Plot para comparar precio según corte
sns.violinplot(x="cut", y="price", data=df, palette="muted")
plt.title("Distribución del Precio por Tipo de Corte")
plt.xlabel("Tipo de Corte")
plt.ylabel("Precio")
plt.show()
```

# Escribe una conclusión basada en los gráficos.

En base a los gráficos, se observa que existe una fuerte relación entre el precio y carat. Puede verse la relación de las variables carat, x, y y la variable z en pairplot. Mientra que el heatmap confirma esto, ya que sus coeficientes superan los valores de 0.85. Cuanto mayor es el tamaño del diamante, mayor es su precio.
En el gráfico violin plot, podemos concluir que  eltipo de corte no determina directamente el precio, puede vverse una gran dispersión en cada categoría. Se observa que en el caso de un corte ideal,tambien presenta precios altos, por lo cual no influye de manera directa en el precio como sí lo hace el tamaño.
