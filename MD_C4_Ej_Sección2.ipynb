```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Usa Matplotlib para crear un gráfico de barras que muestre el promedio de ingresos de diferentes categorías de empleados.

```python
# Crear DataFrame
df = pd.DataFrame({
    "Categoría":        ["Junior", "Semi-Senior", "Senior"],
    "Salario_Promedio": [35000, 60000, 90000]
})

# Gráfico de barras con Matplotlib
plt.bar(df["Categoría"], df["Salario_Promedio"], color=["blue", "green", "red"])
plt.xlabel("Categoría de Empleado")
plt.ylabel("Salario Promedio")
plt.title("Salario Promedio por Categoría")
plt.show()
```

# Usa Seaborn para crear un scatterplot que muestre la relación entre Edad e Ingresos en un dataset de empleados.

```python
# Crear DataFrame con edades e ingresos
df2 = pd.DataFrame({
    "Edad":     [25, 32, 40, 35, 50, 45, 29, 37, 41, 60],
    "Ingresos": [30000, 45000, 60000, 55000, 80000, 75000, 40000, 58000, 72000, 90000]
})

# Scatterplot con Seaborn
sns.scatterplot(x=df2["Edad"], y=df2["Ingresos"])
plt.title("Relación entre Edad e Ingresos")
plt.show()
```

# Explica cuál de los dos gráficos es más fácil de interpretar y por qué.


Considero que el gráfico de barras es más facil de interpretar. A simple vista podemos ver los valores máximos y mínimos, las escalas de cada eje son claras y la altura de cada barra nos permite observar su valor correspondiente en el eje Y de forma rápida y exacta.
